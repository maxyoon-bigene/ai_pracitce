```python
#학번: 60191672 이름:윤태균
import numpy as np
import pandas as pd

print("#1==========================================================================")
dep = ['cs','cs','dt','dcd','dt','as','as','ice','ice','ice']
sno = ['0001','0002','0003','0004','0005','0006','0007','0008','0009','0010']
gpa =[2.10,3.50,4.50,2.70,3.00,3.15,4.00,0.00,3.25,3.70]
print(dep,"\n",sno,"\n",gpa) 
#table 단순 리스트화
print("===================================")
list_b=np.array([['cs','cs','dt','dcd','dt','as','as','ice','ice','ice'],
['0001','0002','0003','0004','0005','0006','0007','0008','0009','0010'],
[2.10,3.50,4.50,2.70,3.00,3.15,4.00,0.00,3.25,3.70]])
print(list_b)
#numpy arrar를 이용해 리스트화
print("===================================")
list_c=pd.Series([dep,sno,gpa],index=['dep','sno','gpa'])
print(list_c)
#pandas series를 이용해 리스트화
print("#2==========================================================================")
gpa_mean_a=round(sum(gpa)/len(gpa),2) #gpa의 평균값 구하기 - 합을 개수로 나누고 두자릿수 출력
print("mean of gpa with python internal libraries: ",gpa_mean_a)
print("===================================")
np.set_printoptions(formatter={'float_kind':lambda x:"{0:0.2f}".format(x)}) #numpy출력값 자릿수 2자리로 설정
gpa_mean_b=np.mean(gpa)  #numpy 내장함수 mean()사용
gpa_median_b = np.median(gpa)  #numpy 내장함수 median()사용
print("median of gpa with numpy: ",gpa_median_b)
print("mean of gpa with numpy: ",gpa_mean_b)
print("===================================")
# pd.options.display.float_format = '{:.2f}'.format
my_dic ={'dep':dep,'sno':sno,'gpa':gpa}  #딕셔너리화
df=pd.DataFrame(my_dic)   #pandas DataFrame 만들기
print("median of gpa with pandas : ",round(df['gpa'].median(),2))  #DataFrame 내장함수 median()사용, round로 두 자릿수 출력
print("mean of gpa with pandas : ",round(df['gpa'].mean(),2))   #DataFrame 내장함수 mean()사용, round로 두 자릿수 출력

print("#3==========================================================================")
def calc_median(lis):  #중간값 구하는 함수 
    lis.sort()         #인자로 받은 lis 리스트 정렬
    med_idx = len(lis)//2   #중간 인덱스값 저장
    med = lis[med_idx]      #med 에 중간값 저장
    return med             #med return
print("median value of gpa is ", calc_median(gpa))

print("#4==============================================================")
df_gpa= df.sort_values(by='gpa',ascending=False)      #gpa 기준으로 내림차순순
print("row with the 3rd highest gpa :\n",df_gpa[2:3])  #gpa기준 3번째로 높은 열 출력

print("#5==============================================================")
#각 dep별로 나뉘어 gpa의 평군, 최대, 최소값을 출력한다
print("======dep가 cs인 그룹=======")
df_cs =df[df.dep =='cs']
# print(df_cs)
print("gpa의 평균 / 최대 / 최소 : ", round(df_cs['gpa'].mean(),2)," / ",
        round(df_cs['gpa'].max(),2)," / ",round(df_cs['gpa'].min(),2))
print("======dep가 dt인 그룹=======")
df_dt =df[df.dep =='dt']
# print(df_dt)
print("gpa의 평균 / 최대 / 최소 : ", round(df_dt['gpa'].mean(),2)," / ",
        round(df_dt['gpa'].max(),2)," / ",round(df_dt['gpa'].min(),2))
print("======dep가 dcd인 그룹=======")
df_dcd =df[df.dep =='dcd']
# print(df_dcd)
print("gpa의 평균 / 최대 / 최소 : ", round(df_dcd['gpa'].mean(),2)," / ",
        round(df_dcd['gpa'].max(),2)," / ",round(df_dcd['gpa'].min(),2))
print("======dep가 as인 그룹=======")
df_as =df[df.dep =='as']
# print(df_as)
print("gpa의 평균 / 최대 / 최소 : ", round(df_as['gpa'].mean(),2)," / ",
        round(df_as['gpa'].max(),2)," / ",round(df_as['gpa'].min(),2))
print("======dep가 ice인 그룹=======")
df_ice = df[df.dep =='ice']
# print(df_ice)
print("gpa의 평균 / 최대 / 최소 : ", round(df_ice['gpa'].mean(),2)," / ",
        round(df_ice['gpa'].max(),2)," / ",round(df_ice['gpa'].min(),2))

print("#6==========================================================================")
df1 = pd.DataFrame({ 
 'dep' : ["cs", "cs", "dt", "dt", "dt", "as", "as", "cs", "dt", "as"],
 'sno' : ["0001", "0002", "0003", "0004", "0005", "0006", "0007", "0008", "0009", "0010"],
 'gpa' : [2.10, 3.50, 4.50, 2.70, 3.00, 3.20, 4.00, 0.00, 3.20, 3.70],
 'year' : [2018, 2018, 2019, 2019, 2020, 2020, 2020, 2020, 2021, 2021]})
#주어진 데이터를 dataframe으로 만들기
dff = pd.pivot_table(df1,index ='dep',columns='year',values='gpa',aggfunc='mean')
"""pivot table이용해서 행 위치에 들어갈 열 dep, 열 위치에 들어갈 열 year, 데이터로 사용할 열 gpa, 
데이터의 집계함수를 mean으로 데이터의 평균값을 출력"""
print(dff)


```
